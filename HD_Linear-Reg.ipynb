3. BUILDING MODEL

```python
x=heart[['Sex', 'age', 'currentSmoker', 'cigsPerDay', 'BPMeds','prevalentStroke', 'prevalentHyp','diabetes','totChol',
'sysBP','diaBP', 'BMI', 'heartRate', 'glucose']]
y=heart['TenYearCHD']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=7)
```

```python

from sklearn.linear_model import LogisticRegression
```

```python
log=LogisticRegression()
```

```python
log.fit(x_train,y_train)
```

```python
pred=log.predict(x_test)
```

```python
h2=pd.DataFrame({"predicted":pred,"actual":y_test})
h2
```

4. PERFORMANCE EVALUATION

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,pred)
```

```python
TP,FN,FP,TN=confusion_matrix(y_test,pred).reshape(-1)
```

```python
TP
```

```python
FN
```

```python
FP
```

```python
TN
```

```python

from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score
```

```python
#Accuracy=(TN+TP)/(TN+TP+FN+FP)
accuracy_score(y_test,pred)
```

```python
#Precision=(TP)/(TP+FN)
precision_score(y_test,pred)
```

```python
#Recall=(TP)/(TP+FP)
recall_score(y_test,pred)
```

```python
#F1-score=(2*Recall*Precision)/(Recall+Precision)
f1_score(y_test,pred)
```

CONCLUSION:

Here the Accuracy is above 0.75 or 75%. So, the Heart Disease Machine Learning Mode can be considered for usage, but Model can be modeified further for the better accuracy and other values.
